# importing needed libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
```

# reading csv data


```python
df = pd.read_csv('marketing_campaign.csv', sep='\t')
print(f"rows =   {df.shape[0]}")
print(f"columns =   {df.shape[1]}")
df.head()
```

```python
data_copy = df.copy()
data_copy.head()
```

# decribe df


```python
df.describe()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df = df[~df['Marital_Status'].isin(['Alone', 'Absurd', 'YOLO'])]
df.reset_index(drop=True, inplace=True)
```

```python
drop_columns = ['Z_CostContact', 'Z_Revenue', 'ID', 'Complain', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Response']
df = df.drop(drop_columns, axis=1)
```

# barchart for education

```python
colors = ['red', 'blue', 'green', 'purple', 'yellow']
ax = sns.barplot(x=my.index,y=my.values, palette=colors)
sns.despine(left=True, bottom=True)
for i in range(len(colors)):
    ax.bar_label(ax.containers[i])
plt.yticks([], []);
```
